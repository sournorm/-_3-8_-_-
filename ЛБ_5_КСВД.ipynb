# 1. Iris

1. 3D-графік розподілу даних "sepal_length", "sepal_width", "petal_length" з розфарбуванням за класом квітки.
2. Гістограма розподілу "petal_width" для різних видів.
3. Лінійний графік зміни середнього "sepal_length" від "sepal_width".

```python
import plotly.express as px

iris_data = px.data.iris()

# ==========================================
# 1.1 Побудова 3D-графіка
# ==========================================
scatter_3d_fig = px.scatter_3d(
    data_frame=iris_data,
    x='sepal_length',
    y='sepal_width',
    z='petal_length',
    color='species',
    title='Тривимірний розподіл характеристик Iris'
)
scatter_3d_fig.update_layout(title={'x': 0.5, 'xanchor': 'center'})
scatter_3d_fig.show()

# ==========================================
# 1.2 Побудова гістограми розподілу "petal_width"
# ==========================================
hist_fig = px.histogram(
    data_frame=iris_data,
    x='petal_width',
    color='species',
    title='Розподіл ширини пелюстки (petal_width) за видами',
)
hist_fig.update_layout(barmode='overlay', title_x=0.5)
hist_fig.show()

# ==========================================
# 1.3 Лінійний графік зміни середнього "sepal_length"
# ==========================================
avg_sepal_df = iris_data.groupby(['species', 'sepal_width']).agg(
    avg_sepal_length=('sepal_length', 'mean')
).reset_index()

line_fig = px.line(
    data_frame=avg_sepal_df,
    x='sepal_width',
    y='avg_sepal_length',
    color='species',
    title='Динаміка середнього sepal_length відносно sepal_width'
)
line_fig.update_traces(mode='lines+markers')
line_fig.update_layout(title_x=0.5)
line_fig.show()
```

# 2. Gapminder
1. Гістограма розподілу топ-10 країн за ВВП у 2007 році .
2. Кругова діаграма розподілу ВВП для 5 країн Європи з найбільшим ВВП .
3. Діаграма динаміки населення Канади та країн Океанії.

```python
#  Gapminder
gapminder_data = px.data.gapminder()

gapminder_data['GDP_Total'] = gapminder_data.eval('gdpPercap * pop')

# ==========================================
# 2.1 Топ-10 країн за ВВП (2007 рік)
# ==========================================
data_2007_top10 = gapminder_data.query("year == 2007").sort_values(by='GDP_Total', ascending=False).head(10)

bar_gdp_fig = px.bar(
    data_frame=data_2007_top10,
    x='country',
    y='GDP_Total',
    color='GDP_Total',
    color_continuous_scale='Plasma',
    title='10 країн з найбільшим обсягом ВВП за 2007 рік',
    text_auto='.2s'
)
bar_gdp_fig.update_traces(textposition='outside')
bar_gdp_fig.update_layout(title_x=0.5, uniformtext_minsize=9, uniformtext_mode='hide')
bar_gdp_fig.show()

# ==========================================
# 2.2 Кругова діаграма: Топ-5 країн Європи за ВВП
# ==========================================
europe_top5_2007 = gapminder_data.query("continent == 'Europe' and year == 2007").sort_values(by='GDP_Total', ascending=False).head(5)

pie_eur_fig = px.pie(
    data_frame=europe_top5_2007,
    names='country',
    values='GDP_Total',
    title='Розподіл ВВП п\'яти найбагатших країн Європи (2007 рік)'
)
pie_eur_fig.update_layout(title_x=0.5)
pie_eur_fig.update_traces(textinfo='percent+label')
pie_eur_fig.show()

# ==========================================
# 2.3 Динаміка населення: Канада та Океанія
# ==========================================
canada_oceania_data = gapminder_data.query("country == 'Canada' or continent == 'Oceania'")

pop_dyn_fig = px.bar(
    data_frame=canada_oceania_data,
    x='year',
    y='pop',
    color='country',
    barmode='group',
    title='Порівняння динаміки населення: Канада та держави Океанії'
)
pop_dyn_fig.update_layout(title_x=0.5)
pop_dyn_fig.show()
```

3. Лінійні графіки
Побудова математичних графіків на заданих проміжках із відцентрованими назвами.

Функція y = 2*cos(x - π/4) на проміжку [0; 5π]
Функція y = 0.5*sin(2x + π/3) на проміжку [0; 2π]
Функція y = 2*tg(0.5x + π/6) на проміжку [-2π; 2π]

```python
import numpy as np
import plotly.express as px

# ==========================================
# 3.1 Графік функції y = 2*cos(x - π/4)
# ==========================================
domain_cos = np.linspace(0, 5 * np.pi, 600)
func_cos = 2 * np.cos(domain_cos - np.pi/4)

# Створимо графік, передаючи словник з назвами осей
fig_cos = px.line(
    x=domain_cos,
    y=func_cos,
    title='Тригонометрична функція: y = 2 * cos(x - π/4)',
    labels={'x': 'Значення X', 'y': 'Значення Y'}
)
fig_cos.update_layout(title={'x': 0.5})
fig_cos.show()


# ==========================================
# 3.2 Графік функції y = 1/2*sin(2x + π/3)
# ==========================================
domain_sin = np.linspace(0, 2 * np.pi, 600)
func_sin = (1/2) * np.sin(2 * domain_sin + np.pi/3)

fig_sin = px.line(
    x=domain_sin,
    y=func_sin,
    title='Тригонометрична функція: y = 1/2 * sin(2x + π/3)'
)
fig_sin.update_traces(line_color='indianred')
fig_sin.update_layout(
    title_x=0.5,
    xaxis_title='Вісь X',
    yaxis_title='Вісь Y'
)
fig_sin.show()


# ==========================================
# 3.3 Графік функції y = 2*tg(1/2*x + π/6)
# ==========================================
domain_tan = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
func_tan = 2 * np.tan(0.5 * domain_tan + np.pi/6)

func_tan[np.abs(func_tan) > 20] = np.nan

fig_tan = px.line(
    x=domain_tan,
    y=func_tan,
    title='Тригонометрична функція: y = 2 * tg(0.5x + π/6)'
)
fig_tan.update_layout(
    title_x=0.5,
    xaxis_title='Вісь X',
    yaxis_title='Вісь Y',
    yaxis=dict(range=[-10, 10])
)
fig_tan.show()
```
