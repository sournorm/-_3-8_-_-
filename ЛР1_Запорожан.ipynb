5. Створіть два одновимірних масиви розміром 6, заповніть їх випадковими цілими числами в діапазоні від 1 до 20 та виконайте на них поелементні операції додавання, віднімання та множення.

```python
arr_x = np.random.randint(1, 21, size=6)
arr_y = np.random.randint(1, 21, size=6)
print(f"Масив X: {arr_x}")
print(f"Масив Y: {arr_y}")
print(f"Результат додавання: {np.add(arr_x, arr_y)}")
print(f"Результат віднімання: {np.subtract(arr_x, arr_y)}")
print(f"Результат множення: {np.multiply(arr_x, arr_y)}")
```

6. Створіть дві матриці розміром 3x2 та 2x4, заповніть їх випадковими цілими числами в діапазоні від 1 до 5 та перемножте їх між собою.

```python
mat_alpha = np.random.randint(1, 6, size=(3, 2))
mat_beta = np.random.randint(1, 6, size=(2, 4))

result_product = mat_alpha @ mat_beta

print("Матриця 3x2:\n", mat_alpha)
print("Матриця 2x4:\n", mat_beta)
print("Результат множення матриць:\n", result_product)
```

7. Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в діапазоні від 0 до 1 та транспонуйте її.

```python
original_m = np.random.rand(3, 3)
transposed_m = np.transpose(original_m)

print("Оригінальна матриця:\n", original_m)
print("Транспонована матриця:\n", transposed_m)
```

8. Створіть дві матриці розміром 3x3, заповніть їх випадковими цілими числами в діапазоні від 1 до 10 та знайдіть їхню різницю.

9. Створіть дві матриці розміром 3x3, заповніть їх випадковими цілими числами в діапазоні від -5 до 5 та знайдіть їх добуток.

```python
m_left = np.random.randint(-5, 6, (3, 3))
m_right = np.random.randint(-5, 6, (3, 3))

final_dot = np.dot(m_left, m_right)
print("Результат добутку (dot):\n", final_dot)
```

10. Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в діапазоні від -1 до 1 та знайдіть вектор-стовпчик, що містить суму елементів кожного рядка матриці.

```python
main_matrix = np.random.uniform(-1, 1, (3, 3))
col_vector_sums = np.sum(main_matrix, axis=1).reshape(-1, 1)

print("Матриця:\n", main_matrix)
print("Вектор-стовпчик сум рядків:\n", col_vector_sums)
```
